```python
import xml.etree.ElementTree as ET
import sys
import carla
sys.path.append("../")
```

```python
# Setup Carla
client = carla.Client("localhost", 2000)
world = client.get_world()
world_map = world.get_map()
```

```python
def parse_element(element):
    result = {}

    for child in element:
        child_result = parse_element(child)

        if child.tag in result:
            if type(result[child.tag]) is list:
                result[child.tag].append(child_result)
            else:
                result[child.tag] = [result[child.tag], child_result]
        else:
            result[child.tag] = child_result

        if not result[child.tag]:
            result[child.tag] = child.text.strip() if child.text else None

        attributes = dict(child.attrib)
        if attributes:
            result[child.tag] = {'text': result[child.tag], **attributes}

    return result
```

```python
# Load Scenario(s)
xosc_file = '/home/lukasdierkes/repos/carla/runnerTool/scenarios/$R0Y4UJA.xosc'
tree = ET.parse(xosc_file)
root = tree.getroot()
parse_element(root)
```

```python
# Performance
performances = root.findall('.//Performance')
print("performances", performances)
# performances[0].attrib["maxSpeed"] = 1000000

# Weather


# TeleportAction
teleport_actions = tree.findall(".//WorldPosition")
# print("teleportAction", teleport_actions)
location = carla.Location(x = float(teleport_actions[0].attrib["x"]), y = float(teleport_actions[0].attrib["y"]), z = float(teleport_actions[0].attrib["z"]))
print("old location", location)
new_location = world_map.get_waypoint(location).next(1)[0].transform.location
teleport_actions[0].attrib["x"] = new_location.x
teleport_actions[0].attrib["y"] = new_location.y
teleport_actions[0].attrib["z"] = new_location.z
print("new_location", new_location)
```
